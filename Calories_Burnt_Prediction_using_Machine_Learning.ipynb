```python
df.replace({'male': 0, 'female': 1},
           inplace=True)
df.head()
```

```python
plt.figure(figsize=(8, 8))
sb.heatmap(df.corr() > 0.9,
           annot=True,
           cbar=False)
plt.show()
```

```python
to_remove = ['Weight', 'Duration']
df.drop(to_remove, axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

features = df.drop(['User_ID', 'Calories'], axis=1)
target = df['Calories'].values

X_train, X_val,\
    Y_train, Y_val = train_test_split(features, target,
                                      test_size=0.1,
                                      random_state=22)
X_train.shape, X_val.shape
```

```python
# Normalizing the features for stable and fast training.
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
from sklearn import metrics
from sklearn.svm import SVC
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error as mae

models = [LinearRegression(), XGBRegressor(),
          Lasso(), RandomForestRegressor(), Ridge()]

for i in range(5):
    models[i].fit(X_train, Y_train)

    print(f'{models[i]} : ')

    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(Y_train, train_preds))

    val_preds = models[i].predict(X_val)
    print('Validation Error : ', mae(Y_val, val_preds))
    print()
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize the model
xgb = XGBRegressor()

# Fit the model on training data
xgb.fit(X_train, Y_train)

# Predict on training and validation data
train_preds = xgb.predict(X_train)
val_preds = xgb.predict(X_val)

# Print Evaluation Metrics
print("XGBoost Model Evaluation")

print("\n--- Training Metrics ---")
print("MAE:", mean_absolute_error(Y_train, train_preds))
print("RMSE:", mean_squared_error(Y_train, train_preds, squared=False))
print("R Score:", r2_score(Y_train, train_preds))

print("\n--- Validation Metrics ---")
print("MAE:", mean_absolute_error(Y_val, val_preds))
print("RMSE:", mean_squared_error(Y_val, val_preds, squared=False))
print("R Score:", r2_score(Y_val, val_preds))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(Y_val, val_preds, color='blue', alpha=0.6)
plt.plot([Y_val.min(), Y_val.max()], [Y_val.min(), Y_val.max()], 'r--')
plt.xlabel('Actual Calories Burnt')
plt.ylabel('Predicted Calories Burnt')
plt.title('Actual vs Predicted Calories (XGBoost)')
plt.grid(True)
plt.show()
```
